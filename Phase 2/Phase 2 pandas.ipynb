```python
'''   
Dataset 1: Employee Data (Classic Real-World Scenario)
import pandas as pd

data = {
    "name": ["Amit", "Sara", "John", "Priya", "David", "Anita"],
    "department": ["IT", "HR", "IT", "Finance", "IT", "HR"],
    "salary": [60000, 45000, 70000, 50000, 80000, 48000],
    "experience": [2, 5, 3, 4, 6, 2],
    "city": ["Delhi", "Mumbai", "Delhi", "Chennai", "Delhi", "Mumbai"]
}

df = pd.DataFrame(data)

👉 Tasks:
    Find the average salary per department
Get employees with salary > 60000 AND experience > 3
Find the highest paid employee in each department
Count employees in each city
Add a new column:
"Senior" if experience ≥ 5, else "Junior"
'''



import pandas as pd

data = {
    "name": ["Amit", "Sara", "John", "Priya", "David", "Anita"],
    "department": ["IT", "HR", "IT", "Finance", "IT", "HR"],
    "salary": [60000, 45000, 70000, 50000, 80000, 48000],
    "experience": [2, 5, 3, 4, 6, 2],
    "city": ["Delhi", "Mumbai", "Delhi", "Chennai", "Delhi", "Mumbai"]
}

df = pd.DataFrame(data)
# print(df)

#1
avg_sal = df.groupby("department")["salary"].mean()
print(f"Average Salary per Department : {avg_sal}")

#2
salary_g_experience_g = df[
    (df["salary"] > 50000) & 
    (df["experience"] > 3)
]
print(f"\nEmployees {salary_g_experience_g}")

#3
highest_paid = df.groupby("department")["salary"].max()
print(f"\nHighest_paid employee in each dept are : {highest_paid}")

#4
count_emp_each_city = df.groupby("city")["name"].count()
print(f"\nCount of employee in each city: {count_emp_each_city}")

#5
df["position"] = df["experience"].apply(lambda x:"Senior" if x>=5 else "Junior")
print(df) 
```

```python
'''
Dataset 2: E-Commerce Orders 🛒
data = {
    "order_id": [101, 102, 103, 104, 105, 106],
    "customer": ["A", "B", "A", "C", "B", "A"],
    "amount": [250, 400, 150, 300, 200, 500],
    "status": ["Delivered", "Cancelled", "Delivered", "Delivered", "Cancelled", "Delivered"]
}

df = pd.DataFrame(data)
👉 Tasks:

Total revenue from delivered orders only
Find customer who spent the most
Count number of cancelled orders
Get average order value per customer
Create a column:
"High Value" if amount > 300 else "Low Value"
'''
import pandas as pd

data = {
    "order_id": [101, 102, 103, 104, 105, 106],
    "customer": ["A", "B", "A", "C", "B", "A"],
    "amount": [250, 400, 150, 300, 200, 500],
    "status": ["Delivered", "Cancelled", "Delivered", "Delivered", "Cancelled", "Delivered"]
}

df = pd.DataFrame(data)
print(df)

#1
revenue_delivered = df[df["status"] == "Delivered"]["amount"].sum()
print(f"\nrevenue_delivered {revenue_delivered}")

#2
most_spent_customer = df.groupby("customer")["amount"].sum().idxmax()
print(f"\nmost_spent_customer {most_spent_customer}")

#3
count_cancelled_orders = df[df["status"] == "Cancelled"].shape[0]
print(f"\ncount_cancelled_orders {count_cancelled_orders}")

#4
avg_order_val_per_customer = df.groupby("customer")["amount"].mean()
print(f"\navg_order_val_per_customer {avg_order_val_per_customer}")

#5
df["value"] = df["amount"].apply(lambda x : "High Value" if x>300 else "Low Value")
print(f"\n{df}")
```

```python
'''
Dataset 3: Student Performance 🎓
data = {
    "student": ["A", "B", "C", "D", "E"],
    "math": [90, 70, 60, 85, 95],
    "science": [85, 75, 65, 80, 90],
    "english": [88, 72, 70, 78, 92]
}

df = pd.DataFrame(data)
👉 Tasks:
Add a column "total_marks"
Add "percentage"
Find the top scorer
Get students scoring > 80 in all subjects
Rank students based on percentage
'''
import pandas as pd


data = {
    "student": ["A", "B", "C", "D", "E"],
    "math": [90, 70, 60, 85, 95],
    "science": [85, 75, 65, 80, 90],
    "english": [88, 72, 70, 78, 92]
}

df = pd.DataFrame(data)

#1
df["total_marks"] = df["math"] + df["science"] + df["english"]
print(f"total_marks \n{df}")

#2
df["percentage"] = ((df["total_marks"]/300) *100)
print(f"percentage \n{df}")



#3
top_student = df.loc[df["percentage"].idxmax()]
print(f"\n{top_student} is the top scorer")

#4
g_80 = df[
    (df["math"] > 80) &
    (df["science"] > 80) &
    (df["english"] > 80)
]

print(f"Students who scored 80+ in all subjects : \n {g_80}")

#5
df["rank"] = df["percentage"].rank(ascending=False).astype(int)
print(df)
```

```python
'''   
Dataset 4: Data Cleaning Scenario 🧹 (VERY IMPORTANT)
data = {
    "name": ["A", "B", "C", "D", "E"],
    "age": [25, None, 30, None, 40],
    "salary": [50000, -999, 60000, -999, 70000]
}

df = pd.DataFrame(data)
👉 Tasks:
Replace -999 with NaN
Fill missing age with mean age
Drop rows where salary is still missing
Convert all columns to correct data types
Detect if any duplicates exist
'''
import pandas as pd
import numpy as np

data = {
    "name": ["A", "B", "C", "D", "E"],
    "age": [25, None, 30, None, 40],
    "salary": [50000, -999, 60000, -999, 70000]
}

df = pd.DataFrame(data)
print(df)

#1
df["salary"] = df["salary"].replace(-999, np.nan)
print(f"\nReplacing -999 with NAN \n {df}")

#2
df["age"] = df["age"].fillna(df["age"].mean())
print(f"\nfilling_missing_age \n{df}")

#3
df = df.dropna(subset=["salary"])
print(f"\nDropped missing values : \n{df}")

#4
df["age"] = df["age"].astype(int)
df["salary"] = df["salary"].astype(int)

print(f"\nAfter resolving type issues: \n{df}")

#5
print(f"\n{df.duplicated()}")
```

```python
'''   
Dataset 6: Real Analyst Thinking (IMPORTANT 🧠)
data = {
    "product": ["A", "B", "C", "A", "B", "C"],
    "region": ["North", "South", "North", "South", "North", "South"],
    "sales": [100, 200, 150, 250, 300, 100]
}

df = pd.DataFrame(data)
👉 Tasks:
Total sales per product
Total sales per region
Create a pivot table
Find best selling product in each region
Normalize sales (0–1 scaling)
'''

import pandas as pd
import numpy as np

data = {
    "product": ["A", "B", "C", "A", "B", "C", "A"],
    "region": ["North", "South", "North", "South", "North", "South", "North"],
    "sales": [100, 200, 150, 250, 300, 100, 200]
}

df = pd.DataFrame(data)

#1
total_sales_per_product = df.groupby("product")["sales"].sum()
print(f"total_sales_per_product: \n{total_sales_per_product}")

#2
total_sales_per_region = df.groupby("region")["sales"].sum()
print(f"\ntotal_sales_per_region: \n{total_sales_per_region}")

#3
pivot = pd.pivot_table(df, values="sales", index="product", columns="region", aggfunc="sum")
print(f"\n{pivot}")

#4
best_selling_product_region = df.loc[df.groupby("region")["sales"].idxmax()]
print(f"\n{best_selling_product_region}")

#5
df["normalized_sales"] = (
    (df["sales"] - df["sales"].min()) /
    (df["sales"].max() - df["sales"].min())
)
print(f"\n{df}")
```

```python
'''Dataset: Superstore Sales (REALISTIC MIX)
import pandas as pd

data = {
    "order_id": [101,102,103,104,105,106,107,108],
    "customer": ["A","B","A","C","B","A","D","C"],
    "region": ["North","South","North","East","South","North","West","East"],
    "product": ["Laptop","Phone","Tablet","Laptop","Tablet","Phone","Laptop","Phone"],
    "category": ["Electronics","Electronics","Electronics","Electronics","Electronics","Electronics","Electronics","Electronics"],
    "sales": [50000,20000,15000,55000,18000,22000,60000,25000],
    "quantity": [2,1,1,3,2,1,4,2],
    "date": pd.to_datetime([
        "2024-01-01","2024-01-02","2024-01-03","2024-01-04",
        "2024-01-05","2024-01-06","2024-01-07","2024-01-08"
    ])
}

df = pd.DataFrame(data)
🧠 1. GroupBy & Aggregation (CORE ANALYST SKILL)
👉 Questions:
Find total sales per region
Find average order value per customer
Get multiple aggregations:
total sales
average quantity per product
Find top-performing region based on revenue
Write a custom aggregation:
“High Value Orders %” per region (sales > 30,000)'''


import pandas as pd

data = {
    "order_id": [101,102,103,104,105,106,107,108],
    "customer": ["A","B","A","C","B","A","D","C"],
    "region": ["North","South","North","East","South","North","West","East"],
    "product": ["Laptop","Phone","Tablet","Laptop","Tablet","Phone","Laptop","Phone"],
    "category": ["Electronics","Electronics","Electronics","Electronics","Electronics","Electronics","Electronics","Electronics"],
    "sales": [50000,20000,15000,55000,18000,22000,60000,25000],
    "quantity": [2,1,1,3,2,1,4,2],
    "date": pd.to_datetime([
        "2024-01-01","2024-01-02","2024-01-03","2024-01-04",
        "2024-01-05","2024-01-06","2024-01-07","2024-01-08"
    ])
}

df = pd.DataFrame(data)
print(df)

#1 Total Sales Per Region
total_sales_per_region = df.groupby("region")["sales"].sum()
print(f"total_sales_per_region : \n{total_sales_per_region}")


#2Find average order value per customer
avg_order_per_value = df.groupby("customer")["sales"].mean()
print(f"\navg_order_per_value : \n{avg_order_per_value}")

#3Get multiple aggregations:
#a)
total_sales = df["sales"].sum()
print(f"\nTotal Sales: \n{total_sales}")
#b)
avg_quantity_per_product = (df.groupby("product")["quantity"].mean()).round(2)
print(f"\nAverage Quantity: \n{avg_quantity_per_product}")

#4 Top-performing region
top_performing_region = df.groupby(["region"])["sales"].sum().idxmax()
print(f"\nTop performing region : \n{top_performing_region}")

#5 Higher Value orders per region
res = df.groupby("region").agg(
    total_sales=("sales", "sum"),
    high_value_pct=("sales", lambda x: (x > 30000).mean() * 100)
)
print(f"\nHigher Value Order: \n{res}")
```

```python
'''   
2. Merging & Joining (REAL BUSINESS SCENARIO)
customers = pd.DataFrame({
    "customer": ["A","B","C","D"],
    "city": ["Delhi","Mumbai","Bangalore","Chennai"]
})
👉 Questions:
Merge sales data with customer city info
Find total sales per city
Perform:
inner join
left join
→ observe difference
Identify customers with no orders (edge case thinking)
'''

import pandas as pd

data = {
    "order_id": [101,102,103,104,105,106,107,108],
    "customer": ["A","B","A","C","B","A","D","C"],
    "region": ["North","South","North","East","South","North","West","East"],
    "product": ["Laptop","Phone","Tablet","Laptop","Tablet","Phone","Laptop","Phone"],
    "category": ["Electronics","Electronics","Electronics","Electronics","Electronics","Electronics","Electronics","Electronics"],
    "sales": [50000,20000,15000,55000,18000,22000,60000,25000],
    "quantity": [2,1,1,3,2,1,4,2],
    "date": pd.to_datetime([
        "2024-01-01","2024-01-02","2024-01-03","2024-01-04",
        "2024-01-05","2024-01-06","2024-01-07","2024-01-08"
    ])
}

customers = pd.DataFrame({
    "customer": ["A","B","C","D"],
    "city": ["Delhi","Mumbai","Bangalore","Chennai"]
})

df1 = pd.DataFrame(data)
df2 = pd.DataFrame(customers)

#1
df3 = pd.merge(df1, df2 ,on="customer", how="left")
print(f"Merged Data : \n{df3}")

#2 Find total sales per city
total_sales_per_city = df3.groupby("city")["sales"].sum()
print(f"\nTotal sales per city: \n{total_sales_per_city}")

#3 Perform inner and left join and observe the difference

inner_join = pd.merge(df1, df2 ,on="customer", how="inner")
print(f"\nInner Join : \n{inner_join}")

left_join = pd.merge(df1, df2 ,on="customer", how="left")
print(f"\nLeft Join : \n{left_join}")

#4 Identify customers with no orders (edge case thinking)
customers_no_orders = (df3["quantity"]==0).sum()
print(f"\nCustomers with no orders : \n{customers_no_orders}")
```
