# Matplotlib in Data Science

```python
'''
📊 Problem 1: Sales Trend Analysis

You are given monthly sales data of a startup:

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales = [12000, 15000, 18000, 17000, 21000, 25000]

👉 Tasks:

Plot a line chart
Add:
Title
Axis labels
Grid
Highlight the highest sales point
Set y-axis limit slightly above max value
'''


import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales = [12000, 15000, 18000, 17000, 21000, 25000]

# Find highest sales
max_sales = max(sales)
max_index = sales.index(max_sales)

# Main line plot
plt.plot(months,sales, color="blue", linestyle="--", marker = "o", label = "Monthly sales")

# Highlight highest point
plt.scatter(months[max_index], sales[max_index], color="red", s=150, label="Highest Sales")

# Labels and styling
plt.title("Graph showing Sales per month ")
plt.xlabel("Months")
plt.ylabel("Sales per month")
plt.ylim(12000,26000)
plt.grid(axis="y")
plt.legend()
plt.show()



```

```python
'''   
Problem 2: Product Comparison (Bar Chart)
products = ["Laptop", "Phone", "Tablet", "Headphones"]
revenue = [50000, 70000, 30000, 20000]

👉 Tasks:

Create a bar chart
Add value labels on top of each bar
Change colors for each bar
Sort bars in descending order before plotting
'''

import matplotlib.pyplot as plt

products = ["Laptop", "Phone", "Tablet", "Headphones"]
revenue = [50000, 70000, 30000, 20000]

combined = list(zip(products,revenue))
sorted_data = sorted(combined, key=lambda x: x[1], reverse = True)

sorted_products, sorted_revenue = zip(*sorted_data)

bars = plt.bar(sorted_products, sorted_revenue, color=["red", "blue", "green", "yellow"])
plt.xlabel("Products")
plt.ylabel("Revenue")
plt.title("Product Revenue Comparison")

plt.grid(axis='y')

# Add value labels on bars
for bar in bars:
    height = bar.get_height()

    plt.text(
        bar.get_x() + bar.get_width()/2,
        height + 1000,
        f"{height}",
        ha='center'
    )
plt.show()


```

```python
''' 
Problem 3: Market Share (Pie Chart)
brands = ["Apple", "Samsung", "Xiaomi", "Others"]
market_share = [30, 25, 20, 25]

👉 Tasks:

Create a pie chart
Show percentage labels
Explode the largest segment
Add shadow + custom colors
'''
import matplotlib.pyplot as plt

brands = ["Apple", "Samsung", "Xiaomi", "Others"]
market_share = [30, 25, 20, 25]

largest_index  = market_share.index(max(market_share))

# Explode only largest slice
explode = [0, 0, 0, 0]
explode[largest_index] = 0.1

plt.pie(x=market_share, labels=brands, autopct='%1.1f%%', explode=explode, shadow=True, colors=["silver", "gold", "lime", "purple"])
plt.legend()
plt.show()
```

```python
'''  
Problem 4: Distribution Check (Histogram)
import numpy as np
data = np.random.normal(50, 10, 1000)

👉 Tasks:

Plot a histogram
Add:
20 bins
Edge color
Identify:
Is data symmetric or skewed?
Draw a vertical line for mean
'''

import matplotlib.pyplot as plt
import numpy as np

data = np.random.normal(50, 10, 1000)
mean_data = np.mean(data)

plt.axvline(mean_data, color="red", linestyle="--", label=f"Mean = {mean_data:.2f}")
plt.hist(data, bins=20, edgecolor = "black")

# Labels
plt.title("Distribution Check")
plt.xlabel("Values")
plt.ylabel("Frequency")

plt.legend()
plt.show()
```

```python
'''  
Problem 5: Relationship Detection (Scatter Plot)
hours_studied = [1,2,3,4,5,6,7,8]
scores = [35,40,50,55,65,70,80,90]

👉 Tasks:

Create a scatter plot
Add best-fit trend line
Change marker style
Interpret relationship (positive/negative)
'''

import matplotlib.pyplot as plt
import numpy as np

hours_studied = [1,2,3,4,5,6,7,8]
scores = [35,40,50,55,65,70,80,90]

plt.scatter(hours_studied, scores, marker="s", color="red", label="Student Data")

#best-fit line
m, b = np.polyfit(hours_studied, scores, 1)

#trend line
trend_line = [m*x + b for x in hours_studied]

#Plotting trend line
plt.plot(hours_studied, trend_line, color="blue", linestyle="--", label="Trend Line")

#Labels
plt.title("Relationship between Hours Study & Scores")
plt.xlabel("hours_studied")
plt.ylabel("scores")
plt.legend()
plt.grid()
plt.show()
```

```python
'''  
Problem 6: Data Quality Check (CRITICAL INDUSTRY SKILL)
ages = [22, 25, 30, 35, 120, 28, 27, 200, 26]

👉 Tasks:

Plot histogram
Detect outliers visually
Highlight suspicious values
Explain why visualization helped here
'''

import matplotlib.pyplot as plt

ages = [22, 25, 30, 35, 120, 28, 27, 200, 26]

plt.hist(ages, bins=10, edgecolor="black", color="skyblue")

# Highlight suspicious values
plt.axvline(120, color="red", linestyle = "--", label="Possible Outlier: 120")
plt.axvline(200, color="darkred", linestyle = "--", label="Outlier: 200")

plt.title("Data Quality Check")
plt.xlabel("ages")
plt.ylabel("frequency")

plt.legend()
plt.grid(axis="y")
plt.show()
```

```python
'''  
Problem 7: Multi-Plot Dashboard (OO Approach REQUIRED)
days = ["Mon", "Tue", "Wed", "Thu", "Fri"]
sales = [200, 220, 250, 270, 300]
profit = [20, 25, 30, 35, 40]

👉 Tasks:

Create 2 subplots:
Sales (line chart)
Profit (bar chart)
Use Object-Oriented API (fig, ax)
Add titles to each subplot
Add overall figure title
'''
import matplotlib.pyplot as plt

days = ["Mon", "Tue", "Wed", "Thu", "Fri"]
sales = [200, 220, 250, 270, 300]
profit = [20, 25, 30, 35, 40]

# Create a grid with 1 row and 2 columns
fig, axes = plt.subplots(1,2, figsize = (10,5))


axes[0].plot(days, sales, marker="o", linestyle = "--", label="Sales Trend", color="blue")
axes[0].set_title("Sales Trend")
axes[0].set_xlabel("Days")
axes[0].set_ylabel("Sales")
axes[0].grid()

axes[1].bar(days, profit, color=["red", "blue", "green", "yellow", "pink"], label="Sales Trend")
axes[1].set_title("Profit Analysis")
axes[1].set_xlabel("Days")
axes[1].set_ylabel("Profit")

fig.suptitle("Company Dashboard", fontsize = 16)

plt.tight_layout()

plt.show()
```
