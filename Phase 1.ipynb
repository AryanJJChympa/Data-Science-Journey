```python
''' 1. Sales Summary (Variables + Loops)

You’re given daily sales:

sales = [120, 340, 88, 210, 450, 60]

👉 Tasks:

Calculate total sales
Find average sale
Count how many days had sales > 200
Print result using f-strings
'''
```

```python
import numpy as np

sales = [120, 340, 88, 210, 450, 60]

Total_sales = 0
no_of_sales = 0
for sale in sales:
    Total_sales+=sale
    no_of_sales+=1
print("Total Sales", Total_sales)

#---------------------------------------------#

Average_Sale = Total_sales/no_of_sales
print(Average_Sale)

#---------------------------------------------#

Sales_G_T_200 = 0
for sale in sales:
    if sale>=200:
        Sales_G_T_200+=1
print(Sales_G_T_200)
```

```python
'''
2. User Data Cleaner (Functions)

You receive messy input:

prices = ["$1,200", "$340", "$99.99", "$5,000"]

👉 Tasks:

Write a function clean_price()
Convert all values to float
Return cleaned list
'''
```

```python
def clean_price(prices_dist):
    cleaned = []
    for price in prices_dist:
        cleaned.append(price.replace("$","").replace(",",""))
    return cleaned
        


prices = ["$1,200", "$340", "$99.99", "$5,000"]
clean_price(prices)
```

```python
'''
3. Login System (Control Flow)

Simulate a login:

correct_password = "data123"

👉 Tasks:

Ask user input (use input())
Allow max 3 attempts
Print:
"Access Granted"
"Account Locked"
'''
```

```python

correct_password = "data123"
attempts = 3

while attempts > 0:
    entered_password = input("Enter the password: ")

    if(entered_password == correct_password):
        print("Access Granted")
        break
    else:
        attempts-=1
        print(f"Access Denied. No. of attempts left: {attempts}")
if attempts == 0:
    print("Out of attempts. Try again later.") 
```

```python
'''
4. Customer Insights (Dict + Loop)
customers = [
    {"name": "Alice", "city": "Delhi", "spend": 5000},
    {"name": "Bob", "city": "Mumbai", "spend": 3000},
    {"name": "Carol", "city": "Delhi", "spend": 7000}
]

👉 Tasks:

Total spend per city
Find highest spender
Create new list of customers with spend > 4000
'''
```

```python
def sum_spent(customers, city):
    total = 0
    for customer in customers:
        if customer["city"] == city:
            total+=customer["spend"]
    return total



customers = [
    {"name": "Alice", "city": "Delhi", "spend": 5000},
    {"name": "Bob", "city": "Mumbai", "spend": 3000},
    {"name": "Carol", "city": "Delhi", "spend": 7000}
]
city = input("Enter the city: ")

print(f"The total spent of {city} : {sum_spent(customers,city)}")


#-------------------------Find Highest Spender--------------------------------#

Highest_spent = 0
Top_Customer = ""

for customer in customers:
        if customer["spend"] > Highest_spent:
            Highest_spent = customer["spend"]
            Top_Customer = customer["name"]

print("Top Customer : ", Top_Customer)
print("Amount Spend: ",Highest_spent)


#-------------------------Create a new list with customer who spent greater than = to 4000--------------------------------#

new_Customer = []

for customer in customers:
        if customer["spend"] >=4000:
             new_Customer.append(customer["name"])
print(new_Customer)
```
